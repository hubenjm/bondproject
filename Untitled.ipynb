```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import visualize
```

```python
x = np.random.rand(100) - 0.5
y = np.random.rand(100) - 0.5
```

```python
cnf_matrix = confusion_matrix((y>0).astype(np.int), (x>0).astype(np.int))
```

```python
fig = visualize.plot_confusion_matrix(cnf_matrix, classes = [0,1], fontsize = 24, tick_fontsize = 16, cmap = plt.cm.YlGn)
```

```python
plt.show()
```
